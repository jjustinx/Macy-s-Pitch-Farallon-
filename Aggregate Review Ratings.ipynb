```python
import json
from collections import defaultdict
from textblob import TextBlob
stores = ["macys","neimanmarcus","gap","dillards","jcpenney","kohls","nordstrom","target","tjmaxx", "bluemercury","sephora","ulta","kiehls","aesop","lush","origins"]

for i, name in enumerate(stores):
    file_name = name+"_reviews.txt"
    with open(file_name) as f:
        reviews = json.load(f)
    data = []
    ratings = []
    for d in reviews:
        #print(d['text'])
        rating = float(d['stars'])
        ratings.append(rating)
    #print(sentiment)
    file_name = name+"_ratings.txt"
    with open(file_name, 'w') as outfile:
        for r in ratings:
            string = str(r) + "\n"
            outfile.write(string)
   # result = ml.classifiers.classify(model_id, data)
   # print(result.body)
```

```python
from matplotlib import pyplot as plt
import numpy as np
 
stores = ["macys","neimanmarcus","gap","dillards","jcpenney","kohls","nordstrom","target","tjmaxx", "bluemercury","sephora","ulta","kiehls","aesop","lush","origins"]
for i, name in enumerate(stores):
    file_name = name+"_ratings.txt"
    ratings = []
    with open(file_name) as f:
        for line in f:
            ratings.append(float(line))
# Creating dataset
    a = np.array(ratings)
    print(stores[i] + " median rating: ", np.average(a))
    
    fig, ax = plt.subplots(figsize =(10, 7))
    ax.hist(a, bins = [0,1,2,3,4,5])
    plt.title(stores[i])
    #plt.show()
    
```
